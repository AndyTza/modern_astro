```python
import h5py as hp
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hf = hp.File('/Users/iraf1/Downloads/gaia_spectro.hdf5', 'r')
```

```python
l = np.array(list((hf['data']['l'])))
l = l[~np.isnan(l)]
b = np.array(list((hf['data']['b'])))
b = b[~np.isnan(b)]
```

```python
%config InlineBackend.figure_format = 'svg'

l_bins = np.arange(0,360, step=0.4)
b_bins = np.arange(-80, 80, step=0.4)

H = np.histogram2d(np.degrees(b), np.degrees(l), bins=(b_bins, l_bins))

_, ax = plt.subplots(nrows=1, ncols=1, figsize=(10,5))
ax.imshow(np.log(H[0]), cmap='coolwarm', origin='lower', aspect='auto',
          extent=[0,360, -80, 80], interpolation='nearest', vmin=0, vmax=5)
plt.xlabel('$\mathcal{\ell}$ [deg]', size=15)
plt.ylabel('$b$ [deg]', size=15)
```
